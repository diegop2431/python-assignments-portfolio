```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
print ("Ready")
```

```python
N0= 10000  #initial population
I0= 2  #initial infected
R0= 0  #initial recovered 
D0= 0  #initial dead
S0=N0-I0-R0-D0 #intial susceptible
y0_sird =  N0, S0, I0, R0, D0
t = np.linspace(0, 100, 100)

beta= 0.8 #rate of infection 
gamma= 0.1 #rate of recovery
omega= 0.05 #per capita growth rate
delta= 0.3 #per capita death rate of infected
roh= 0.001 #per capita natural death rate
def model(y, t, beta, gamma, omega, delta, roh):
    N, S, I, R, D = y
    dNdt= omega*N - delta*I - roh*N
    dSdt= -beta*S*I/N - roh*S + omega*N
    dIdt= beta*S*I/N - gamma*I - delta*I-roh*I
    dRdt= gamma*I-roh*R
    dDdt= delta*I+roh*N
    return [dNdt, dSdt, dIdt, dRdt, dDdt]

ret_sird = odeint(model, y0_sird, t, args=(beta, gamma, omega, delta, roh))
N_sird, S_sird, I_sird, R_sird, D_sird = ret_sird.T

plt.figure(figsize=(10, 6))
plt.plot(t, N_sird, 'teal', label='Population') 
plt.plot(t, S_sird, 'b', label='Susceptible')
plt.plot(t, I_sird, 'r', label='Infected')
plt.plot(t, R_sird, 'g', label='Recovered')
plt.plot(t, D_sird, 'k', label='Deceased')
plt.title('SIRD Model')
plt.xlabel('Time / days')
plt.ylabel('Number of Individuals')
plt.legend()
plt.grid(True)
plt.show()
```
