```python
a =162
b =190
n = 1000
x = np.linspace (a, b, n)
mean = 171
variance = 7.1

def normal_density(mean, variance, x):
    return (1/np.sqrt(2*np.pi*variance))*np.exp((-(x-mean)**2)/2*variance)
plt.plot (x, normal_density(mean, variance, x))
plt.show

def integration(mean, variance, a, b):
    deltax=(b-a)/n
    return deltax*(1/np.sqrt(2*np.pi*variance))*np.exp((-(x-mean)**2)/2*variance)
print (np.sum(integration(mean, variance, a, b)))
```
