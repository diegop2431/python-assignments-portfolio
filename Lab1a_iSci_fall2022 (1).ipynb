<h3> Intermediate Value Theorem </h3>


<b> Intermediate Value Theorem, IVT: </b>
If $f(x)$ is a continuous function defined on a closed interval $[a,b]$ and $N$ is a number between $f(a)$ and $f(b)$, then there is a number $c$ in $[a,b]$ such that $f(c)=N$.

<i> In words:</i> A continuous function $f(x)$ defined on an interval $[a,b]$ must assume all values between $f(a)$ and $f(b).$

For example, consider the air temperature, which is a continuous function of time. Assume that at 9am it was measured to be ${16}^{\hbox{o}}\hbox{C}$, and at 11am it was ${21}^{\hbox{o}}\hbox{C}$. By the IVT, we can conclude that between 9am and 11am, the temperature had to assume every value between ${16}^{\hbox{o}}\hbox{C}$ and ${21}^{\hbox{o}}\hbox{C}$. So for instance, at some time between 9am and 11am, it had to be ${18.4}^{\hbox{o}}\hbox{C}$. 

In symbols: Let $f(t)$=temperature at time $t.$ We know that $f(9)=16$ and $f(11)=21.$ Let $N=18.4.$ Because $N$ is between $f(9)=16$ and $f(11)=21,$ it had to happen! I.e., there is some time $c$ in the interval $[9,11],$ when $f(c)=18.4.$




<b> Special case (most often used): </b>
Take $N=0,$ i.e., we are solving the equation $f(x)=0$ for some continuous function $f(x).$ If we find an interval $[a,b]$ so that at one end $f(x)$ is positive, and at the other it is negative, then there must be a point in $[a,b]$ such that $f(c)=0,$ i.e., $c$ is a solution.



<b> Example: </b>

Use the Intermediate Value Theorem to show that the equation $e^x = 5x+10$ has a solution for $3 \leq x \leq 4$. 

<i> Solution: </i> Let $f(x)=e^x-5x-10$ and $N=0$. We calculate that $f(3) \approx -4.9$ and $f(4) \approx 24.6$. Since $f(x)$ is continuous on $[3,4]$ and $N=0$ is between $f(3)$ and $f(4)$, the Intermediate Value Theorem guarantees that there is a number $c$ in the interval $[3,4]$ such that $f(c)=0$. See below.



```python
# So we know that f(x)=0 for some x between 3 and 4. How do we make a better estimate?

a = 3
b = 4
n = 10

x = np.linspace(a,b,n+1)
y = np.exp(x) - 5*x - 10

for i in range (len(x)):
    print("(", x[i], ",", y[i], ")")
    
# Run this cell
```

<h3> Using the Intermediate Value Theorem in Python </h3>

Run the code in the cell below to see another way Python can be used to answer this question. 

In the code above, we had to search through the list of $y$-values to find the change in sign. How does the code below locate the change of sign for us?

```python
## Show that y=e^x-5x-10=0 has a solution on [a,b]

a = 3
b = 4
n = 10

x = np.linspace(a,b,n+1)
y = np.exp(x) - 5*x - 10
        
for i in range (1, n+1):
    if y[i-1] == 0.: # check if y=0 at the left end of a small interval; if so, we found a solution
        print("One solution of f(x)=0 is", x[i-1])
    if y[i-1] * y[i] < 0.:  # check if a solution is inside the interval
        print("f(x)=0 has a solution between", x[i-1], "and", x[i]) 
    if y[i] == 0.:  # check if y=0 at the right end of a small interval; if so, we found a solution
        print("One solution of f(x)=0 is", x[i]) 
        
plt.plot(x,y)
plt.grid()
plt.show()

# Run this code
```

<i> Summary: </i>

Using Python, we generate a table of $n$ values for a function $f(x)$ and compute the product of all adjacent pairs of $y$-values, that is, we compute $y[i-1]*y[i]$ for all $1 \leq
i\leq n$. 

If the product is positive for a pair of values, $y[i-1]$ and $y[i]$, then both values have the same sign, that is, they are either both negative or both positive. In this case, we do not know - the function might, or might not cross the $x$-axis:



If the product is negative for a pair of values, $y[i-1]$ and $y[i]$, then the $y$-values have different signs, that is, one is negative and the other is positive. For a continuous function, we can conclude that $f(x)$ must have taken on the value of 0 somewhere between $x[i-1]$ and $x[i]$, that is $f(c)=0$ for some number $c$ between $x[i-1]$ and $x[i]$. 

Of course, if the product $y[i-1]*y[i]=0$, then either $y[i-1]=0$ or $y[i]=0$ which implies that the solution is $c=x[i-1]$ or $c=x[i]$.

<h3> Differentiation </h3>

The derivative of $f(x)$ with respect to $x$ is the function $\displaystyle{f'(x) = \lim_{h\to 0} {f(x+h) - f(x) \over h} }$.

The average rate of change of $f(x)$ on the interval $[a, a+h]$ is given by $\displaystyle{ {f(a+h) - f(a) \over h} }$. This difference quotient can be used to approximate the derivative of $f(x)$ at $x=a$. 

```python
# estimating the derivative of f(x)=x^2 on [a,b]

a = -5  # as usual, a and b define the interval on which we study the given function 
b = 5
n = 200
h = 0.01 # step-size

# define variables
x = np.linspace(a,b,n)
y = x**2
y_prime = ((x+h)**2 - x**2)/h  # IMPORTANT: note that we did not enter y_prime = 2x !!

# plot functions
plt.plot(x,y, label = '$f(x)$', color = 'green')
plt.plot(x,y_prime, label = "$f'(x)$")
plt.title("Graphs of $f(x)$ and $f'(x)$")
plt.legend()
plt.grid()
plt.show()

# run the cell ... does the derivative graph make sense?
```

```python
# estimating the derivative of f(x)= square root of x on [a,b]

a = 0.1
b = 4
n = 100
h = 0.01 # step-size

# define variables
x = np.linspace(a,b,n)
y = x**(1/2)  # square root function
y_prime = ((x+h)**(1/2) - x**(1/2))/h  # as in the previous cell, we did not use the formula for y'

# plot functions
plt.plot(x,y, label = '$f(x)$', color = 'green')
plt.plot(x,y_prime, label = "$f'(x)$")
plt.title("Graphs of $f(x)$ and $f'(x)$")
plt.legend()
plt.grid()
plt.show()

# does the graph of f'(x) make sense?
```
