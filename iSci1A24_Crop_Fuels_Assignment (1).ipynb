 <h3><b> (2) Plant Biomass Function  </b></h3>

In this section, we will use Euler's method to approximate the crop growth (plant biomass) function, $PBM(sq)$. To begin our analysis, we will consider two limited growth models: the logistic model and the Gompertz function, both of which may be used to describe the crop growth rate with respect to soil quality (per hectare) for each of our crops.

<u>**Logistic Model**</u>

$\displaystyle{ {dPBM \over dsq} = kPBM\left(1-{PBM \over L}\right)}$

where $k$ is a positive constant and $L$ is the carrying capacity.

<u>**The Gompertz Function**</u>
 
The Gompertz function is a solution of 

$\displaystyle{ {dPBM \over dsq} = c\ln\left({M \over PBM}\right)PBM}$

where $c$ is a constant and $M$ is the carrying capacity.

```python
## LOGISTIC MODEL FOR PBM (code for 2(a))

# parameters, variables
n = 20 # number of iterations/steps
h = 0.5  # step size
L = 10.53 # defining variables
k = 1
x = [0]    #initializing the x value in the set
y = [0.5]  #initializing the y value in the set

# euler's algorithm
for i in range(1, n+1):
    xi = x[i-1] + h
    yi = y[i-1] + (k*(y[i-1])*(1-(y[i-1]/L)))*h # the logistic model with defined variables
    x.append(xi) #adding a new calculated x value to the set for every loop iteration
    y.append(yi) #adding a new calculated y value to the set for every loop iteration
   
# code for plot
plt.plot(x,y)
plt.xlabel("sq") #labeling x-axis
plt.ylabel("PBM") #labeling y-axis
plt.title("Logistic Approximation of PBM") #adding a title
plt.grid() #adding a grid
# output statement, such as "The approximate value of PBM(10) is: "
print ("The approximate value of PBM(10) is:", y[20])
```

<h3> (4) <b> Total Energy Balance </b> </h3>

Total energy balance is defined as the ratio of human energy out to human energy in:

$\displaystyle{TEB(sq) = {E_o(sq) \over E_i(sq)} }$

where

<b> Energy in</b>: $E_i(sq) = E_S(sq) + C + P_c\cdot PBM(sq)+(E_S(sq)-E_S(sq-D_{sq}))$

<b> Energy out</b>: $E_o(sq) = F_{ED}\cdot PBM(sq)$

<i> * <u> Note</u>: </i> $C$, $P_c$, $D_{sq}$, and $F_{ED}$ are all given in Table 1 near the beginning of this file.

```python
## optimizing TEB (code for 4, parts (a) and (b))

# parameters, variables
sq = np.linspace(0, 10, 500) #creating a set of 500 sq values, equally spaced between 0 and 10
M = 10.53 #defining variables
c = 1
k = -3.047
C = 0.0344
Pc = 2.9
Dsq = 0.13
Fed = 26.8
# formula for TEB
def TEB(sq):
    return (Fed*(M*np.exp(k*np.exp(-c*sq))))/(((sq**3)/10+0.5)+C+(Pc*((sq**3)/10+0.5))+(((sq**3)/10+0.5)-(((sq-Dsq)**3)/10+0.5))) 
# code for plot
plt.plot(sq, TEB(sq)) 
plt.xlabel("sq") #labeling x axis
plt.ylabel("TEB") #labeling y axis
plt.title("Total Energy Balance as a Function of sq") #adding a title
plt.grid() #adding a grid
plt.show()
# locating maximum value, output statement such as "The maximum value of TEB is ___ and occurs when sq is ___ ."
MAX = max(TEB(sq)) #locating the maximum TEB value in set

#locating sq value that gives maximum TEB value
for sq in sq: #iterating through set of sq values
    if TEB(sq) == MAX: #testing if TEB of each sq value equals the maximum TEB value
        print ("The maximum value of TEB is", MAX,"and occurs when sq is", sq) #outputting maximum TEB value and sq value when maximum TEB occurs
```
